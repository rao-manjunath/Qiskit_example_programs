```python
%matplotlib inline
# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, execute, Aer, IBMQ
from qiskit.compiler import transpile, assemble
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *

# Loading your IBM Q account(s)
provider = IBMQ.load_account()
```

```python
from qiskit.optimization import QuadraticProgram
from qiskit.aqua import aqua_globals, QuantumInstance
from qiskit.optimization.algorithms import MinimumEigenOptimizer
from qiskit.aqua.algorithms import NumPyMinimumEigensolver
```

```python
mod = QuadraticProgram('box_pair')
print(mod.export_as_lp_string())
```

```python
mod.binary_var(name='x')
mod.binary_var(name='y')
mod.binary_var(name='z')
print(mod.export_as_lp_string())
```

```python
mod.minimize(linear={'x':17, 'y':21, 'z':19})
print(mod.export_as_lp_string())
```

```python
mod.linear_constraint(linear={'x': 1, 'y': 1, 'z': 1}, 
                         sense='==', 
                         rhs=2, 
                         name='lin_eq_1')
print(mod.export_as_lp_string())
```

```python
print(mod.export_as_lp_string())
```

```python
backend = Aer.get_backend('statevector_simulator')
```

```python
e_mes = NumPyMinimumEigensolver()
```

```python
exact = MinimumEigenOptimizer(e_mes)
```

```python
e_res = exact.solve(mod)
```

```python
print(e_res)
```
